## Reconstruct a String from its Genome Path
http://rosalind.info/problems/ba3b/

```python
with open("rosalind_ba3b.txt", "r") as f:
        k_mer_strings = [line.strip() for line in f]
print(k_mer_strings)
```

```python
first_k_mer = k_mer_strings[0]
for i in range(1,len(k_mer_strings)):
    last_letter = k_mer_strings[i][-1]
    first_k_mer = first_k_mer +  last_letter
print(first_k_mer)
```
