## Find the Minimum Number of Coins Needed to Make Change
http://rosalind.info/problems/ba5a/

```python
%%time
coins = [1,5,10,20,25,50]
total =40
infinite_table = [0]+[float('inf')]*total
print("Table:",infinite_table)

for coin in coins:
    print("Coin = ",coin)
    for i in range(coin,total+1):
        print("i value = ",i,"Table[i] = ", infinite_table[i], "table[i-coin]+1 = ",infinite_table[i-coin]+1, end = " ")
        infinite_table[i]=min(infinite_table[i],infinite_table[i-coin]+1)
        print("Final table[i] = ",infinite_table[i],"Final table",infinite_table)
        
print(infinite_table[total])
```
